```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10)
```

```python
y=x+np.cos(x)
```

```python
plt.axhline(y=0,color="r",linestyle="-")
plt.plot(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
x=np.arange(0,10)
```

```python
y=2*x+1
```

```python
plt.plot(x,y)
```

```python
interp_func=interp1d(x,y)
```

```python
new_x=np.arange(3,6,0.1)
new_x
```

```python
new_y=interp_func(new_x)
```

```python
plt.plot(new_x,new_y)
```
